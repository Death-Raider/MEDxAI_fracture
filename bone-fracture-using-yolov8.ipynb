```python
labels = []
for file in test_image:
    spliting = file.split("/")
    label_path = "/".join(spliting[:-2]+["labels"] + [spliting[-1][:-3]+"txt"])
    with open(label_path) as f:
        data = f.read()
        cls = data.split(" ")[0]
        if cls == '':
            labels.append(-1)
        else:
            labels.append(int(cls))
print(labels)
```
